**Parte 1:** Clase Base Robot

```python
class Robot:
    def __init__(self, name, model):
        self.name = name
        self.model = model
        self.battery = 100.0
        self.distance_traveled = 0.0
        self.speed = 0.0
        self.direction = 0

    def __str__(self):
        return f"{self.name} ({self.model}) - Batería: {self.battery}% - Distancia: {self.distance_traveled:.2f}km"

    def move(self, speed, direction, time_minutes):
        # Actualizar velocidad y dirección
        self.speed = speed
        self.direction = direction

        # Calcular distancia recorrida
        distancia = speed * (time_minutes / 60)
        self.distance_traveled += distancia

        # Reducir batería (10% por km recorrido)
        self.battery -= distancia * 10

        # Alerta cuando batería es menor a 20%
        if self.battery < 20:
            print("Batería baja!")

    def charge(self, time_minutes):
        # Formula de aumento del nivel de la batería
        self.battery += time_minutes * 2

        # Limitar a máximo 100%
        if self.battery > 100:
            self.battery = 100

        print(f"{self.name} cargando... Batería: {self.battery}%")

    def get_status(self):
        print(f"Estado De robot {self.name}:")
        print(f"Modelo: {self.model}")
        print(f"Batería: {self.battery}%")
        print(f"Distancia recorrida: {self.distance_traveled} km")
        print(f"Velocidad actual: {self.speed} km/h")
        print(f"Dirección: {self.direction}°")
```

**Parte 2:** Herencia - Tipos de Robots

```python
class WorkerRobot(Robot):
    def __init__(self, name, model):
        super().__init__(name, model)
        self.cargo = 0.0

    def move(self, speed, direction, time_minutes):
        # Actualizar velocidad y dirección
        self.speed = speed
        self.direction = direction

        # distancia recorrida
        distancia = speed * (time_minutes / 60)
        self.distance_traveled += distancia

        # reducir batería (10% por km recorrido)
        consumo_bateria = distancia * 10

        # Si tiene carga, consumir 20% más batería
        if self.cargo > 0:
            consumo_bateria = consumo_bateria *1.2

        self.battery -= consumo_bateria

        # Verificar si la batería es menor a 20%
        if self.battery < 20:
            print("Batería baja!")

    def load_cargo(self, weight):
        self.cargo += weight
        print(f"Cargando {weight}kg. Total: {self.cargo}kg")

    def unload_cargo(self):
        self.cargo = 0
        print("Carga descargada")
    # El Robot de la clase padre no tiene el atrinuto cargo
    def get_status(self):
        super().get_status()
        print(f"Carga actual: {self.cargo}kg")


class GuardRobot(Robot):
    def __init__(self, name, model):
        super().__init__(name, model)
        self.alert_mode = False

    def charge(self, time_minutes):
        cargaBot =  time_minutes * 2
        # Si está en modo alerta, cargar el robot 50% más lento
        if self.alert_mode:
             self.battery +=cargaBot*0.5
        else:
             self.battery += cargaBot

        # Limitar a máximo 100%
        if self.battery > 100:
            self.battery = 100

        print(f"{self.name} cargando... Batería: {self.battery}%")

    def toggle_alert(self):
        # Negación para invertir el estado de alerta del bot
        self.alert_mode = not self.alert_mode
        if self.alert_mode:
            print("Modo alerta ON")
        else:
            print("Modo alerta OFF")

    def get_status(self):
        super().get_status()
        if self.alert_mode:
            print("Modo alerta activado")
        else:
            print("Modo alerta desactivado")

```

**Parte 3:** Composición - Clase RobotTeam

```python
class RobotTeam:
    def __init__(self, team_name):
        self.team_name = team_name
        # Atributo de instancia, ya que los equipos no comparten los mismos robots
        self.robots = []
        self.missions_completed = 0
        print(f"Equipo {team_name} formado")

    def add_robot(self, robot):
        self.robots.append(robot)
        print(f"{robot.name} se unió al equipo {self.team_name}")

    def __len__(self):
        return len(self.robots)

    def charge_all(self, time_minutes):
        # Carga de los robots usando el método charge de cada uno
        for robot in self.robots:
            robot.charge(time_minutes)

    def team_status(self):
        print(f"Estado del equipo {self.team_name}")
        print(f"Total de robots: {len(self)}")
        print(f"Misiones completadas: {self.missions_completed}")


    def complete_mission(self):
        self.missions_completed += 1
        print(f"¡Misión completada! Total: {self.missions_completed}")
```

**Parte 4:** Código de Prueba

```python
def probarBots():
    print("INICIO DE PRUEBA DE LOS ROBOTS")


    # robots básicos
    print("Creando robots básicos")
    robot1 = Robot("Otto", "1997-292")
    robot2 = Robot("WallE", "2008-4")

    # WorkerRobot
    print("Creando WorkerRobot")
    worker = WorkerRobot("Carguero", "C-3000")

    # GuardRobot
    print("Creando GuardRobot")
    guard = GuardRobot("Guardián", "v-4000")

    # PRUEBA DE MOVIMIENTO

    print("Moviendo robot Otto")
    robot1.move(40, 30, 15)    # 40 km/h, 30°, 5 min

    print("Moviendo robot WallE")
    robot2.move(25, 180, 7)   # 25 km/h, 180°, 7 min

    print("Moviendo robot Carguero")
    worker.move(15, 90, 10)   # 15 km/h, 90°, 10 min

    print("Moviendo robot Vigilante")
    guard.move(35, 270, 3)    # 35 km/h, 270°, 3 min

    # FUNCIONES ESPECIALES
    print("Cargando peso en WorkerRobot")
    worker.load_cargo(50)
    worker.load_cargo(3)

    print("Activando modo alerta en GuardRobot")
    guard.toggle_alert()       # Cambiar modo alerta

    # Carga de las baterías

    print("Cargando robot Otto")
    robot1.charge(2)

    print(" Cargando robot WallE ")
    robot2.charge(10)

    print(" Cargando robot Carguero")
    worker.charge(15)

    print("Cargando robot Vigilante")
    guard.charge(9)

    # ESTADO INDIVIDUAL
    print("Estado de Otto")
    robot1.get_status()

    print("Estado de WallE")
    robot2.get_status()

    print("Estado de Carguero")
    worker.get_status()

    print("Estado de Vigilante")
    guard.get_status()

    # GESTIÓN DE EQUIPOS

    print("Creando equipo de prueba")
    equipo = RobotTeam("Guardianes")

    print("Agregar robots al equipo")
    equipo.add_robot(robot1)
    equipo.add_robot(robot2)
    equipo.add_robot(worker)
    equipo.add_robot(guard)

    # Método len() personalizado
    print(f"Contando robots en el equipo")
    print(f"El equipo '{equipo.team_name}' tiene {len(equipo)} robots")

    # Cargar robots del equipo
    print("Carga colectiva del equipo")
    equipo.charge_all(3)

    # Mostrar status del equipo
    print("Estado completo del equipo")
    equipo.team_status()

    # Completar una misión
    print("Completando misiones")
    equipo.complete_mission()
    equipo.complete_mission()

    # Estado final del equipo
    print("Estado final del equipo")
    equipo.team_status()



probarBots()
```
